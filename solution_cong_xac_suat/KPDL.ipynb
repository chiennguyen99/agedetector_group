```python
import numpy as np
from group import *
```

```python
def model(train_file, test_file, result_file=None):
    groups = parse_group(train_file)
    
    def predict_label(test):
        result = {}
        random_res = ['__label__18-24', '__label__25-34', '__label__35-44', '__label__45-54', '__label__55+']
        for group_id in test:
            if group_id in groups:
                for key in groups[group_id]:
                    if key != 'total':
                        if key in result:
                            result[key] += groups[group_id][key]
                        else: 
                            result[key] = groups[group_id][key]
        res = ""
        max = 0
        if result:
            for key in result:
                if max<result[key]:
                    max = result[key]
                    res = key
        else:
            res = random_res[np.random.randint(5)]
        return res  
    
    def predict(test_file):
        test = read_test_not_label(test_file)
        res = [predict_label(x) for x in test]
        return res 
    
    def read_label_from_data(result_file):
        lines = read_data_from_file(result_file)
        labels = []
        for line in lines:
            labels.append(line.split(" ", 1)[0])
        return labels
    
    def cal_acc(test, result): 
        res = 0
        n = len(test)
        for i in range(len(test)):
            if test[i] == result[i]:
                res += 1
        return res/n
    
    test = predict(test_file)
    if result_file:  
        result = read_label_from_data(result_file)
        return test, cal_acc(test, result)
    return test
```

```python
def save_file(labels, out_file):
    with open(out_file, 'w') as f:
        for label in labels:
            f.write("{}\n".format(label))
```

```python
for i in range(1,6):
    ftrain = "./data_new/test_{}/data_train.txt".format(i)
    ftest = "./data_new/test_{}/data_test.txt".format(i)
    fresult = "./data_new/test_{}/result.txt".format(i)
    fsave = "./data_new/test_{}/test_result.txt".format(i)
    test, acc = model(ftrain, ftest, fresult)
    save_file(test, fsave)
    print("Testcase", i, ": ",acc)
```
